Series を {'a':1,'b':2,'c':3} から作成し index 指定。

```python
import pandas as pd
import numpy as np
```

```python
d={'a':1,'b':2,'c':3}
#シリーズを作成
pd_series=pd.Series(d)
#インデックスから値を取得
print(pd_series['c'])
#DFを作るときはリストから作成。辞書型からは作れない
df=pd.DataFrame([d])
df
```

```python
#乱数で行列作成
A=np.array([[np.random.randint(0,10) for _ in range(10)] for _ in range(3)]).reshape(10,3)
#DFに変換、カラム名をｘ，ｙ，ｚに
df=pd.DataFrame(A,columns=['x','y','z'])
df
```
